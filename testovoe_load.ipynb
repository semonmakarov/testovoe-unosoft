```python
import os
import numpy as np
from osgeo import gdal
import gdown

# Ссылки на файлы TIF на Google Диске
file_urls = [
    "https://drive.google.com/uc?id=1FuH8UqgaGPUdPdG_E_8DMxlmz7LJgbDa",
    "https://drive.google.com/uc?id=1ayNETXkJ9NM9NGGmbvWr8KsHrcHvZmlN",
    "https://drive.google.com/uc?id=10Mfr4HTybCQUk0jlCwUA6b6O_vJ0dMhH",
    "https://drive.google.com/uc?id=1xHC5rgppK225E_uN1VhFy7fRK68mynr0"
]

# Загрузка файлов TIF
output_folder = 'downloaded_tif_files'
os.makedirs(output_folder, exist_ok=True)

# Загружаем файлы и проверяем их наличие
downloaded_files = []
for i, url in enumerate(file_urls):
    output_file = os.path.join(output_folder, f"file_{i + 1}.tif")
    try:
        gdown.download(url, output_file, quiet=False)
        if os.path.exists(output_file):
            downloaded_files.append(output_file)
            print(f"Файл {output_file} успешно загружен")
        else:
            print(f"Ошибка: файл {output_file} не был создан")
    except Exception as e:
        print(f"Ошибка при загрузке файла {i+1}: {str(e)}")

```

```python
def split_tif_to_tiles(input_file, output_folder, tile_size=1280):
    from osgeo import gdal
    ds = gdal.Open(input_file)
    if ds is None:
        return False

    width = ds.RasterXSize
    height = ds.RasterYSize

    for i in range(0, width, tile_size):
        for j in range(0, height, tile_size):
            w = min(tile_size, width - i)
            h = min(tile_size, height - j)

            gdal.Translate(
                f"{output_folder}/tile_{i}_{j}.tif",
                ds,
                srcWin=[i, j, w, h],
                format='GTiff'
            )
    return True


# Разбиение загруженных TIF файлов на тайлы
tiles_output_folder = 'tiles'
os.makedirs(tiles_output_folder, exist_ok=True)

# Проверяем наличие загруженных файлов
if not downloaded_files:
    print("Нет файлов для разбиения на тайлы")
else:
    for tif_file in downloaded_files:
        print(f"\nОбработка файла: {tif_file}")
        success = split_tif_to_tiles(tif_file, tiles_output_folder)
        if success:
            print(f"Файл {tif_file} успешно разбит на тайлы")
        else:
            print(f"Ошибка при разбиении файла {tif_file}")
```

```python
import os
import numpy as np
from osgeo import gdal
import gdown
from PIL import Image
import math

def convert_tif_to_png_with_resize(input_file, output_file, max_size=16000):
    """Конвертирует TIF в PNG с учетом ограничений размера"""
    try:
        with Image.open(input_file) as img:
            # Получаем размеры изображения
            width, height = img.size

            # Проверяем, нужно ли уменьшить изображение
            if width > max_size or height > max_size:
                # Вычисляем коэффициент масштабирования
                scale = min(max_size / width, max_size / height)
                new_width = int(width * scale)
                new_height = int(height * scale)

                # Изменяем размер изображения
                img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

            # Конвертируем в RGB если нужно
            if img.mode in ['RGBA', 'LA'] or (img.mode == 'P' and 'transparency' in img.info):
                img = img.convert('RGB')

            # Сохраняем как PNG
            img.save(output_file, 'PNG', optimize=True)
            print(f"Успешно конвертирован файл: {os.path.basename(input_file)}")
            return True
    except Exception as e:
        print(f"Ошибка при конвертации {input_file}: {str(e)}")
        return False

def process_tif_files(input_folder, batch_size=10):
    """Обрабатывает TIF файлы партиями"""
    # Создаем папку для PNG файлов
    png_folder = 'tiles_png'
    os.makedirs(png_folder, exist_ok=True)

    # Получаем список всех TIF файлов
    tif_files = [f for f in os.listdir(input_folder) if f.endswith(('.tif', '.tiff'))]

    # Обрабатываем файлы партиями
    for i in range(0, len(tif_files), batch_size):
        batch = tif_files[i:i + batch_size]
        print(f"\nОбработка партии {i//batch_size + 1}/{math.ceil(len(tif_files)/batch_size)}")

        for tif_file in batch:
            tif_path = os.path.join(input_folder, tif_file)
            png_path = os.path.join(png_folder, tif_file.replace('.tif', '.png').replace('.tiff', '.png'))

            if not os.path.exists(png_path):  # Пропускаем уже сконвертированные файлы
                convert_tif_to_png_with_resize(tif_path, png_path)

    return png_folder

def upload_to_roboflow(png_folder, project, batch_size=5):
    """Загружает PNG файлы в Roboflow партиями"""
    png_files = [f for f in os.listdir(png_folder) if f.endswith('.png')]

    for i in range(0, len(png_files), batch_size):
        batch = png_files[i:i + batch_size]
        print(f"\nЗагрузка партии {i//batch_size + 1}/{math.ceil(len(png_files)/batch_size)}")

        for png_file in batch:
            try:
                file_path = os.path.join(png_folder, png_file)
                project.upload(file_path, batch_name="batch_" + str(i//batch_size + 1))
                print(f"Успешно загружен файл: {png_file}")
            except Exception as e:
                print(f"Ошибка при загрузке {png_file}: {str(e)}")

# Основной код
if __name__ == "__main__":
    input_folder = 'tiles'  # Папка с исходными TIF файлами

    # Обработка TIF файлов
    print("Начало конвертации TIF в PNG...")
    png_folder = process_tif_files(input_folder)

    # Загрузка в Roboflow
    print("\nНачало загрузки файлов в Roboflow...")
    try:
        upload_to_roboflow(png_folder, project)
        print("\nЗагрузка завершена. Теперь выполните разметку в интерфейсе Roboflow.")
    except Exception as e:
        print(f"\nОшибка при загрузке в Roboflow: {str(e)}")
```

```python
from roboflow import Roboflow
import json

def read_roboflow_config(config_file='config.txt'):
    with open(config_file, 'r') as f:
        config = json.load(f)
    return config

def init_roboflow():
    config = read_roboflow_config()
    rf = Roboflow(api_key=config['api_key'])
    project = rf.workspace(config['workspace']).project(config['project'])
    return project

read_roboflow_config("/content/config.txt")
init_roboflow()
```

```python
# Загрузка PNG файлов в Roboflow
for png_file in os.listdir(png_folder):
    if png_file.endswith('.png'):
        try:
            file_path = os.path.join(png_folder, png_file)
            project.upload(file_path)
            print(f"Успешно загружен файл: {png_file}")
        except Exception as e:
            print(f"Ошибка при загрузке {png_file}: {str(e)}")

print("\nФайлы загружены в Roboflow. Теперь выполните разметку в интерфейсе Roboflow.")
```
